```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data= pd.read_csv("Student.csv")
```

```python
print("Dataset Info:")
print(data.info())

print("Summary:")
print(data.describe())

print("Mean of Maths Marks:")
print(data['Marks_Math'].mean())
print("Maximum Science Marks:")
print(data['Marks_Science'].max())
```

```python
more_than_50 = data[data['Marks_Math'] > 50]  
print("Students with Maths > 50:", len(more_than_50))
highest_science = data.loc[data['Marks_Science'].idxmax()]  
print("Student with Highest Science Marks:\n", highest_science)
correlation = data['Marks_Math'].corr(data['Marks_Science'])
print("Correlation between Maths and Science:", correlation)
```

```python
plt.hist(data['Age'], bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Histogram: Age Distribution")
plt.show()
```

```python
plt.bar(data['Student_ID'], data['Marks_Math'])
plt.xlabel("Student ID")
plt.ylabel("Math marks")
plt.title("Math marks by Student")
plt.show()
```

```python
plt.scatter(data['Marks_Math'], data['Marks_Science'], color='green')
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title("Scatter Plot: Math vs Science")
plt.show()
```
