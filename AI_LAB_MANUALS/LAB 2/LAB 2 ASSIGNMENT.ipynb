```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("house.csv")
X = df[["Size_sqft", "Bedrooms", "Age_years"]]
y = df["House_Price"]
model = LinearRegression()
model.fit(X, y)
```

```python
print("Intercept:", model.intercept_)
print("Coefficients:", list(zip(X.columns, model.coef_)))

sample = [[2000, 3, 10]]
pred_price = model.predict(sample)
print("Predicted house price:", round(pred_price[0], 2))
```

```python
#Q2
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("student_performance.csv")
X = df[["Hours_Study", "Hours_Sleep", "Attendance_pct"]]
y = df["Marks"]

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
print("R^2:", r2_score(y, y_pred))
print("MSE:", mean_squared_error(y, y_pred))
```

```python
plt.scatter(y, y_pred)
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.title("Actual vs Predicted Marks")
plt.plot([0,100],[0,100], linestyle="--")  # ideal line
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pass_fail.csv")
X = df[["Hours_Study", "Hours_Sleep"]]
y = df["Pass"]

model = LogisticRegression()
model.fit(X, y)
```

```python
sample = [[30, 6]]
prob = model.predict_proba(sample)[0][1]
print("Probability of passing:", round(prob, 3))

# Decision boundary plot (simple)
# create grid
h1 = np.linspace(X["Hours_Study"].min()-1, X["Hours_Study"].max()+1, 200)
h2 = np.linspace(X["Hours_Sleep"].min()-1, X["Hours_Sleep"].max()+1, 200)
xx, yy = np.meshgrid(h1, h2)
grid = np.c_[xx.ravel(), yy.ravel()]
probs = model.predict_proba(grid)[:,1].reshape(xx.shape)

```

```python
plt.contourf(xx, yy, probs, levels=[0,0.5,1], alpha=0.2)
plt.scatter(X["Hours_Study"], X["Hours_Sleep"], c=y, edgecolor='k')
plt.xlabel("Hours Study")
plt.ylabel("Hours Sleep")
plt.title("Decision boundary (Pass probability)")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
df = pd.read_csv("diabetes.csv")
X = df[["BMI", "Age", "Glucose"]]
y = df["Diabetic"]

model = LogisticRegression(max_iter=1000)
model.fit(X, y)
y_pred = model.predict(X)
```

```python
print("Accuracy:", accuracy_score(y, y_pred))
print("Precision:", precision_score(y, y_pred))
print("Recall:", recall_score(y, y_pred))

sample = [[28, 45, 150]]
pred = model.predict(sample)[0]
prob = model.predict_proba(sample)[0][1]
print("Predicted label (1=diabetic):", int(pred), "Probability:",round(prob,3))
```
