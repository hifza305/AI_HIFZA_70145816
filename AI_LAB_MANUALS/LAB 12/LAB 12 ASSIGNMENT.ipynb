```python
import gymnasium as gym
import pygame
import numpy as np

# Q2: Environment setup
env = gym.make("CartPole-v1", render_mode="human")

font = None
episode_scores = [] # Task 4: Store scores

# Task 1: Modify episodes from 20 to 50
for episode in range(1, 51):
    score = 0
    state, info = env.reset() # Q3: Returns state vector and info dictionary
    done = False

    while not done:
        # Task 8: Rule-Based Action (Advanced)
        # Replacing action = env.action_space.sample() (Q6)
        if state[2] > 0:
            action = 1 # Push Right
        else:
            action = 0 # Push Left
            
        # Task 7: Identify State Variables
        # state contains: [Cart Position, Cart Velocity, Pole Angle, Pole Angular Velocity]

        state, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated # Q4: Difference between terminated and truncated
        score += reward # Q5: Calculate total reward

        # Task 5: Slow Down Visualization
        pygame.time.delay(20)

        # Pygame Visualization logic
        if font is None:
            pygame.font.init()
            font = pygame.font.SysFont("Arial", 24)

        surface = pygame.display.get_surface()
        # Task 2 & 3: Display Episode/Score, Color Green (0,255,0), Position (200,20)
        display_text = f"Episode: {episode} | Score: {int(score)}"
        text = font.render(display_text, True, (0, 255, 0)) 
        surface.blit(text, (200, 20)) 
        
        # Q8: Update display (Removing this would freeze the visuals)
        pygame.display.update()

    episode_scores.append(score)
    print(f"Episode {episode} Score: {score}")

# Task 4: Print Maximum Score
print(f"\nMaximum Score Achieved: {max(episode_scores)}")

env.close()
pygame.quit()
```

```python
import gymnasium as gym
import pygame

# Q2: Setup MountainCar Environment
env = gym.make("MountainCar-v0", render_mode="human")

font = None
best_score = -float('inf')
all_scores = [] # Task 4: Track best performance

# Task 1: Set to 50 episodes
NUM_EPISODES = 50 

for episode in range(1, NUM_EPISODES + 1):
    state, info = env.reset()
    done = False
    score = 0
    steps = 0 # Task 8: Track steps per episode

    while not done:
        # Task 2 & 7: Print State Values (Position and Velocity)
        # state[0] = Position, state[1] = Velocity
        
        # Task 7: Simple Rule-Based Policy
        if state[1] > 0:
            action = 2 # Push Right
        else:
            action = 0 # Push Left
            
        state, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated
        score += reward
        steps += 1

        # Task 5: Slow Down Visualization
        pygame.time.delay(20)

        if font is None:
            pygame.font.init()
            font = pygame.font.SysFont("Arial", 24)

        surface = pygame.display.get_surface()
        # Task 3: Change color to Blue (0,0,255) and Position (200,20)
        text = font.render(f"Ep: {episode}  Score: {int(score)}", True, (0, 0, 255))
        surface.blit(text, (200, 20))
        pygame.display.update()

    all_scores.append(score)
    # Task 8: Print steps and analyze relation with score
    print(f"Episode {episode} Score: {score} | Steps: {steps}")
    
    if score > best_score:
        best_score = score

# Task 4: Print the best (highest) score
print(f"\nBest Score Achieved: {best_score}")

env.close()
pygame.quit()
```
