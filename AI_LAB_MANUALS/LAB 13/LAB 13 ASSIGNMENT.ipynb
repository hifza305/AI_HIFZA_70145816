```python
import os
import gensim
import nltk
import pandas as pd
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
from sklearn.decomposition import PCA
import plotly.express as px
import plotly.io as pio


nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Path to your GoT .txt files
DATA_PATH = r"E:\University\Semster 5\AI Lab\venv\datasets\GoT_Data" 

story = []

if os.path.exists(DATA_PATH):
    files = [f for f in os.listdir(DATA_PATH) if f.endswith(".txt")]
    print(f"Found {len(files)} text files.")

    for filename in files:
        file_path = os.path.join(DATA_PATH, filename)
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                corpus = f.read()
        except UnicodeDecodeError:
            with open(file_path, "r", encoding="cp1252") as f:
                corpus = f.read()

        # Sentence Tokenization
        raw_sentences = sent_tokenize(corpus)
        for sent in raw_sentences:
            cleaned_sent = simple_preprocess(sent)
            if cleaned_sent:
                story.append(cleaned_sent)
    
    print(f"Success: {len(story)} total sentences processed.")
else:
    print("Error: Folder path not found. Check DATA_PATH!")
```

```python
if len(story) > 0:
    # Training the model
    model = gensim.models.Word2Vec(window=10, min_count=2)
    model.build_vocab(story)
    model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
    print("Model Training Complete!")
else:
    print("No data found to train.")
```

```python
print("--- Analysis Results ---")
# Task: Find most similar to Daenerys
print("1. Similar to 'daenerys':", model.wv.most_similar('daenerys'))

# Task: Find the odd one out
print("2. Odd one out:", model.wv.doesnt_match(['jon','robb','arya','sansa','bran','tyrion']))

# Task: Calculate Similarity
print("3. Similarity (Arya vs Sansa):", model.wv.similarity('arya','sansa'))

# Task: Show word vector
print("4. Vector of 'king' (partial):", model.wv['king'][:5])
```
